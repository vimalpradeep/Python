```python
#category the people by their age like children, adult, citizen, senior citizen
a= int(input("Age:"))
if(a<18):
    print("Children")
elif(a<35):
    print("Adult")
elif(a<59):
    print("Citizen")
else:
    print("Senior Citizen")
```
